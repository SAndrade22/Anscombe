### 1. Cargar el dataset de Anscombe

```python
import seaborn as sns
import pandas as pd

# Cargar el dataset de Anscombe desde seaborn
anscombe = sns.load_dataset("anscombe")

# Mostrar los primeros datos para confirmar la carga
print(anscombe.head())
```

### 2. Resúmenes estadísticos de Anscombe

```python
# Resúmenes estadísticos del cuarteto de Anscombe
summary_anscombe = anscombe.groupby('dataset').describe()

# Mostrar los resúmenes estadísticos
print(summary_anscombe)
```

### 3. Graficar los datasets de Anscombe

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Graficar los datasets de Anscombe
sns.lmplot(x="x", y="y", col="dataset", hue="dataset", data=anscombe, 
           col_wrap=2, ci=None, palette="muted", height=4, 
           scatter_kws={"s": 50, "alpha": 1})
plt.show()
```

### 4. Graficar los Box-Plot de los datasets de Anscombe

```python
# Graficar Box-Plot de los datasets de Anscombe
sns.boxplot(x="dataset", y="y", data=anscombe)
plt.show()
```
